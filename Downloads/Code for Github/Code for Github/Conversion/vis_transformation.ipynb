## Packages

```python
def readFile(inputPath,fileName):
    inputFile = os.path.join(inputPath, fileName)
    if not os.path.exists(inputFile):
        return fileName
    reader = vtk.vtkPolyDataReader()
    reader.SetFileName(inputFile)
    reader.Update()
    polydata = reader.GetOutput()
    return polydata
```

```python
#allows to store VTKs for later use
class store(object):
    def __init__(self, number):
        self.number = number
```

```python
INPUT_PATH = "/Users/williamengelhardt/OneDrive/Yale_Lab_Summer/data/"
INPUT_PATH_RIGID = os.path.join(INPUT_PATH,"linear_to_ref")
INPUT_PATH_NONRIGID = os.path.join(INPUT_PATH,"nonlinear_from_ref")
```

```python
VTK_base = []
VTK_rigid = []
VTK_nonrigid = []
count = 0
pre = ['a','m','t','y']
for l in pre:
    for i in range(1,50):

        rigid = readFile(INPUT_PATH_RIGID,"%s_%d_total_pre.vtk" %(l,i))
        if isinstance(rigid,str):
            print("%s could not be found" %rigid)
            continue
        nonRigid = readFile(INPUT_PATH_NONRIGID,"%s_%d_total_trans_pre.vtk" %(l,i))
        if isinstance(nonRigid,str):
            print("%s could not be found" %nonRigid)
            continue

        #print('Looking at patient %d: Base image %d and Follow Up image %d'
        #      %(row['patient number'], row['baseline_image_number'], row['follow_up_image_number']))

        
        VTK_rigid.append(store(rigid))
        VTK_nonrigid.append(store(nonRigid))

        count += 1
print('Finished reading %d patients this run' % count)
    
```
